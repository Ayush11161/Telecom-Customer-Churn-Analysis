```python
df["Churn"] = df["Churn"].map({'Yes':1, 'No':0})
```

```python
df.head()
```

```python
df['TenureGroup'] = pd.cut(df['tenure'],
                           bins=[0,12,24,48,60,100],
                           labels=['0-1yr','1-2yr','2-4yr','4-5yr','5+yr'])
```

```python
df['MonthlyGroup'] = pd.cut(df['MonthlyCharges'],
                           bins=[0,35,70,120],
                           labels=['Low','Medium','High'])
```

```python
df.head()
```

```python
df['Churn'].mean()
```

```python
df.groupby('Contract')['Churn'].mean()
```

```python
df.groupby('TenureGroup', observed=True)['Churn'].mean()
```

```python
df.groupby('MonthlyGroup')['Churn'].mean()
```

```python
df.groupby('PaymentMethod')['Churn'].mean()
```

```python
df.groupby('InternetService')['Churn'].mean()
```

```python
df.to_csv("output.csv", index=False)
```
